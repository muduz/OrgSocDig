```python
#importo package
import numpy as np
from matplotlib import pyplot as plt
```

1. **cambiare il colore e la trasparenza del plot, cambiare il titolo**

```python
ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='b', alpha=0.8) #colore e trasparenza

plt.title("Una semplice visualizzazione") #titolo
plt.show()
```

2. **che valori hanno ys, len(ys) e x? (usare print() per vedere il valore)**

```python
print(x) #visualizza i valori
print(ys)
print(len(ys))
```

4. **poi moltiplicare per 10 la variazione casuale di ys**
5. **e sistemare il fill perch√© sia visualizzato correttamente**
